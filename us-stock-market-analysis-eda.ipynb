<b>Crude_oil_Vol.:</b> Trading volume of crude oil.<br>

<b>Bitcoin_Vol.: </b>Trading volume of Bitcoin.<br>

<b>Platinum_Vol.: </b>Trading volume of platinum.<br>
<b>Ethereum_Price: </b>Price of Ethereum in USD.<br>
<b>Ethereum_Vol.: </b>Trading volume of Ethereum.<br>
<b>S&P_500_Price: </b>Price index of the S&P 500.<br>
<b>Nasdaq_100_Price: </b>Price index of the Nasdaq 100.<br>
<b>Nasdaq_100_Vol.: </b>Trading volume for the Nasdaq 100 index.<br>
<b>Apple_Price: </b>Stock price of Apple Inc. in USD.<br>
<b>Apple_Vol.: </b>Trading volume of Apple Inc. stock.<br>
<b>Tesla_Price: </b>Stock price of Tesla Inc. in USD.<br>
<b>Tesla_Vol.: </b>Trading volume of Tesla Inc. stock.<br>
<b>Microsoft_Price:</b> Stock price of Microsoft Corporation in USD.<br>
<b>Microsoft_Vol.: </b>Trading volume of Microsoft Corporation stock.<br>
<b>Silver_Price: </b>Price of silver in USD per troy ounce.<br>
<b>Silver_Vol.: </b>Trading volume of silver.<br>
<b>Google_Price: </b>Stock price of Alphabet Inc. (Google) in USD.<br>
<b>Google_Vol.: </b>Trading volume of Alphabet Inc. stock.<br>
<b>Nvidia_Price: </b>Stock price of Nvidia Corporation in USD.<br>
<b>Nvidia_Vol.: </b>Trading volume of Nvidia Corporation stock.<br>
<b>Berkshire_Price:</b> Stock price of Berkshire Hathaway Inc. in USD.<br>
<b>Berkshire_Vol.: </b>Trading volume of Berkshire Hathaway Inc. stock.<br>
<b>Netflix_Price: </b>Stock price of Netflix Inc. in USD.<br>
<b>Netflix_Vol.: </b>Trading volume of Netflix Inc. stock.<br>
<b>Amazon_Price: </b>Stock price of Amazon.com Inc. in USD.<br>
<b>Amazon_Vol.: </b>Trading volume of Amazon.com Inc. stock.<br>
<b>Meta_Price: </b>Stock price of Meta Platforms, Inc. (formerly Facebook) in USD.<br>
<b>Meta_Vol.: </b>Trading volume of Meta Platforms, Inc. stock.<br>
<b>Gold_Price: </b>Price of gold in USD per troy ounce.<br>
<b>Gold_Vol.: </b>Trading volume of gold.<br>

## Commodities:

```python
commodities = ['Natural_Gas_Price', 'Natural_Gas_Vol.', 
               'Crude_oil_Price', 'Crude_oil_Vol.']
```

```python
# creating a new dataframe for commodities data

commodities_data = df[['Date'] + commodities]
commodities_data
```

Setting date as the index for this df

```python
commodities_data.set_index('Date', inplace=True)
```

```python
commodities_data.describe()
```

Commodities Price trends over time

```python
plt.figure(figsize=(12, 6))

for col in ['Natural_Gas_Price', 'Crude_oil_Price']:
    plt.plot(commodities_data.index, commodities_data[col], label=col)

plt.title('Commodity Price Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

Comparing average trading volumes of commodities

```python
average_commodities_volumes = commodities_data[['Natural_Gas_Vol.', 'Crude_oil_Vol.']].mean()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=average_commodities_volumes.index, y=average_commodities_volumes.values, palette='viridis')
plt.title('Average Trading Volume of Commodities')
plt.xlabel('Commodities')
plt.ylabel('Average Volume')
plt.show()
```

Crude Oil is traded a lot more in volume than Natural Gas

Drawing a correlation matrix:

```python
correlation_commodities_data = commodities_data[['Natural_Gas_Price', 'Natural_Gas_Vol.', 
                                     'Crude_oil_Price', 'Crude_oil_Vol.']]
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_commodities_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Commodities Prices and Volumes')
plt.show()
```

Natural Gas and Crude Oil Price are highly correlated.

```python
palette = { 2019: 'blue', 2020: 'orange', 2021: 'green', 2022: 'red', 2023: 'purple', 2024: 'brown' }
```

## Companies:

```python
tech_companies = ['Apple_Price', 'Apple_Vol.', 
                  'Tesla_Price', 'Tesla_Vol.', 
                  'Microsoft_Price', 'Microsoft_Vol.', 
                  'Google_Price', 'Google_Vol.', 
                  'Nvidia_Price', 'Nvidia_Vol.', 
                  'Berkshire_Price', 'Berkshire_Vol.', 
                  'Netflix_Price', 'Netflix_Vol.', 
                  'Amazon_Price', 'Amazon_Vol.', 
                  'Meta_Price', 'Meta_Vol.']

tech_data = df[['Date'] + tech_companies]

tech_data.set_index('Date', inplace=True)
```

```python
tech_data.describe()
```

FAANG Companies Price trend over time:

```python
plt.figure(figsize=(12, 6))
for col in ['Meta_Price', 'Apple_Price', 'Amazon_Price', 'Netflix_Price', 'Google_Price']:
    plt.plot(tech_data.index, tech_data[col], label=col)

plt.title('Tech Companies Price Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

```python
selected_companies = ['Apple_Price', 'Tesla_Price', 'Microsoft_Price', 'Google_Price', 'Nvidia_Price', 
                      'Meta_Price', 'Netflix_Price', 'Amazon_Price', 'Berkshire_Price']
selected_volumes = [col.replace("Price", "Vol.") for col in selected_companies]

tech_data = df[['Date'] + selected_companies + selected_volumes]
```

```python
tech_data[selected_companies].describe()
```

Calculating average prices and volumes

```python
average_com_prices = tech_data[selected_companies].mean()
average_com_volumes = tech_data[selected_volumes].mean()

# Combined bar plot for average prices and volumes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for average prices
average_com_prices.plot(kind='bar', color='skyblue', ax=ax1, position=0, width=0.4, label='Average Price')
ax1.set_ylabel('Average Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Bar plot for average volumes (on secondary y-axis)
ax2 = ax1.twinx()
average_com_volumes.plot(kind='bar', color='orange', ax=ax2, position=1, width=0.4, label='Average Volume')
ax2.set_ylabel('Average Volume', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Comparison of Average Prices and Volumes')
plt.xticks(rotation=45)
fig.tight_layout()
plt.show()
```

Drawing correlation matrix:

```python
com_price_corr = tech_data[selected_companies].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(com_price_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Company Prices')
plt.show()
```

```python
tech_data.set_index('Date', inplace=True)
```

Stock Price trends for all companies:

```python
# List of companies and their prices
company_prices = ['Apple_Price', 'Tesla_Price', 'Microsoft_Price', 'Google_Price', 'Nvidia_Price',
                  'Meta_Price', 'Netflix_Price', 'Amazon_Price', 'Berkshire_Price']

# Create subplots for time-series trends
num_companies = len(company_prices)
rows = (num_companies + 2) // 3  # Adjust number of rows for subplots
fig, axes = plt.subplots(rows, 3, figsize=(15, rows * 4))

# Flatten axes for easy indexing
axes = axes.flatten()

# Plot each company's time-series data
for idx, company in enumerate(company_prices):
    ax = axes[idx]
    ax.plot(tech_data.index, tech_data[company], label=company.split('_')[0], color='tab:blue')
    ax.set_title(f"{company.split('_')[0]} Price Trend")
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.legend()
    ax.grid()

# Hide any extra subplots (if rows * cols > num_companies)
for i in range(num_companies, len(axes)):
    axes[i].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.suptitle("Time-Series Trends for Tech Companies", fontsize=16, y=1.02)
plt.show()
```

Combined correlation among different categories:

```python
# Combine datasets for correlation analysis
combined_data = pd.concat([
    commodities_data[['Natural_Gas_Price', 'Crude_oil_Price']],
    metals_data[["Gold_Price", "Silver_Price", "Copper_Price"]],
    cryptos_data[['Bitcoin_Price', 'Ethereum_Price']],
    tech_data[selected_companies]
], axis=1)

# Compute correlation matrix
combined_corr = combined_data.corr()

# Heatmap for cross-category correlations
plt.figure(figsize=(12, 8))
sns.heatmap(combined_corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Across Categories')
plt.show()
```

Natural_Gas and Crude_Oil are highly correlated with each other, Companies like Google, Apple, Microsoft, etc. are also highly correlated with each other. These are all expected.
